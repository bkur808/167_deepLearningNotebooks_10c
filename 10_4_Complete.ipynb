```python
# Now let's try max-pooling
def maxpool(x_in):
  x_out = np.zeros(( int(np.floor(x_in.shape[0]/2)), int(np.floor(x_in.shape[1]/2)) ))
  for x in range(x_out.shape[0]):
      for y in range(x_out.shape[1]):
        x_index = 2 * x
        y_index = 2 * y
        # Taking the max value in the 2x2 block
        x_out[x, y] = np.max(x_in[x_index:x_index + 2, y_index:y_index + 2])
  return x_out
```

```python
print("Original:")
print(orig_4_4)
print("Maxpooled:")
print(maxpool(orig_4_4))
```

```python
# Finally, let's try mean pooling
def meanpool(x_in):
  x_out = np.zeros(( int(np.floor(x_in.shape[0]/2)), int(np.floor(x_in.shape[1]/2)) ))
  for x in range(x_out.shape[0]):
      for y in range(x_out.shape[1]):
        x_index = 2 * x
        y_index = 2 * y
        # Taking the max value in the 2x2 block
        x_out[x, y] = np.mean(x_in[x_index:x_index + 2, y_index:y_index + 2])

  return x_out
```

```python
# Define 2 by 2 original patch
orig_2_2 = np.array([[6, 8], [8,4]])
print(orig_2_2)
```

```python
# Let's first use the duplication method
def duplicate(x_in):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  for x in range(x_in.shape[0]):
      for y in range(x_in.shape[1]):
        x_out[2 * x, 2 * y] = x_in[x, y]
        x_out[2 * x + 1, 2 * y] = x_in[x, y]
        x_out[2 * x, 2 * y + 1] = x_in[x, y]
        x_out[2 * x + 1, 2 * y + 1] = x_in[x, y]

  return x_out
```

```python
print("Original:")
print(orig_2_2)
print("Duplicated:")
print(duplicate(orig_2_2))
```

```python
# Now let's try max pooling back up
# The input x_high_res is the original high res image, from which you can deduce the position of the maximum index
def max_unpool(x_in, x_high_res):
  x_out = np.zeros(( x_in.shape[0]*2, x_in.shape[1]*2 ))
  for x in range(x_in.shape[0]):
    for y in range(x_in.shape[1]):
      # Find the 2x2 block in the original high-res image
      block = x_high_res[2 * x:2 * x + 2, 2 * y:2 * y + 2]
            
      # Locate the position of the maximum value in this block
      max_pos = np.unravel_index(np.argmax(block), block.shape)
            
      # Place the value from x_in back into this max position in x_out
      x_out[2 * x + max_pos[0], 2 * y + max_pos[1]] = x_in[x, y]
    
  return x_out

  
```
